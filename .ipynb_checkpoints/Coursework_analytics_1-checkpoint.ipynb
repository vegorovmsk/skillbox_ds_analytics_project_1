```python
# Общее количество курсов в датасете

courses_number = data.course_id.nunique()
courses_number
```

```python
# Количество модулей на каждом курсе

modules_count = data.groupby(['title'])\
                    ['module_number'].nunique()\
                    .reset_index()\
                    .rename(columns={'module_number': 'modules_count'})\
                    .sort_values(by='modules_count', ascending=False)\
                    .reset_index(drop=True)
modules_count
```

```python
# Количество уроков в каждом модуле на каждом курсе

lessons_count = data.groupby(['title', 'module_title'])['lesson_number']\
                        .nunique().reset_index()\
                        .rename(columns={'lesson_number': 'lessons_count'})
lessons_count
```

```python
# медианное количество уроков в модуле на каждом курсе

lessons_median = lessons_count.groupby(['title'])['module_number']\
                                    .agg('median').reset_index()\
                                    .rename(columns={'module_number': 'lessons_median'})
lessons_median
```

```python
# количество учеников на каждом курсе

student_count_by_course = data.groupby(['title'])['student_id']\
                                .nunique().reset_index()\
                                .rename(columns={'student_id': 'students_count'})\
                                .sort_values(by='students_count', ascending=False)\
                                .reset_index(drop=True)
student_count_by_course
```

---
Для работы с датасетом студентов необходимо подготовить данные:

- удалить пустые значения дней рождения;
- отбросить явно неправильные даты (дающие возраст более 100 лет и менее 10 лет);
- конвертировать столбец с днями рождений в формат даты;
- добавить столбец возраста путем вычитания даты рождения из текущей даты.

```python
students_df = students_df.dropna(subset=['birthday']).reset_index(drop=True)
```

```python
students_df.fillna(value='Unknown', inplace=True)
```

```python
students_ages = students_df[(students_df.birthday > '1920-01-01') & (students_df.birthday < '2009-01-01')]
```

```python
students_ages.loc[:, 'birthday'] = students_ages['birthday'].apply(pd.to_datetime)
```

```python
students_ages.loc[:, 'age'] = (pd.to_datetime('today') - students_ages['birthday']) / timedelta(days=365)
```

```python
students_ages
```

```python
# минимальный, максимальный, средний, медианный возраст студентов

age_statistics = students_ages.agg({'age':[min, max, 'mean', 'median']})
age_statistics
```

```python
student_courses_df = students_ages.merge(progresses_df, how='left', left_on='id', right_on='student_id', suffixes=('_stud', '_prog')) 
student_courses_df = student_courses_df.merge(courses_df, how='left', left_on='course_id', right_on='id')
student_courses_df.drop(columns=['id_prog', 'student_id', 'course_id', 'id', 'field'], inplace=True)
student_courses_df.dropna(inplace=True)
student_courses_df
```

```python
# минимальный, максимальный, средний, медианный возраст студентов на каждом курсе

ages_by_courses = student_courses_df.groupby('title')['age'].agg([min, max, 'mean', 'median']).reset_index()
ages_by_courses
```

Цвет столбцов должен содержать информацию о сфере, к которой относится курс (то есть нужна легенда). Прозрачность должна стоять на отметке 0.1. На график должна быть нанесена линия медианы.
```python
# Создадим несколько вспомогательных объектов для группировки bar-chart по сферам, создадим цветовую палитру и список меток легенды.

fields = {v: k for k, v in enumerate(courses_df['field'].unique())}
```

```python
fields_id = student_count_by_course.merge(courses_df, how='left', on='title')['field'].apply(lambda k: fields[k])
```

```python
colors = {0: 'r', 1: 'g', 2: 'b', 3: 'c'}
```

```python
import matplotlib.patches as mpatches
patchlist = [mpatches.Patch(color=col, label=lab) for (col, lab) in zip(colors.values(), fields.keys())]
patchlist.append(mpatches.Patch(color='black', label='Median'))
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
sns.barplot(data=student_count_by_course, x='students_count', palette=[colors[i] for i in fields_id], y='title', alpha=0.1, ax=ax)
sns.despine(ax=ax, left=True, bottom=True)

plt.title('Распределение студентов по курсам', fontsize=16)
plt.legend(handles=patchlist, loc=1, bbox_to_anchor=(1.2, 1))
plt.vlines(student_count_by_course.students_count.median(), ymin=0, ymax=14, colors='black', alpha=0.5)
ax.set_xlabel(None)
ax.set_ylabel(None)
ax.tick_params(labelsize=14)
plt.show()
```

```python
# Необходимо привести значения столбцов начала и окончания учёбы к формату даты. Отбросим значения времени.

data_df['start_date'] = data['start_date'].apply(lambda x: str(x).split(' ')[0])
data_df['start_date'] = data['start_date'].apply(pd.to_datetime)
data_df['finish_date'] = data['finish_date'].apply(lambda x: str(x).split(' ')[0])
data_df['finish_date'] = data['finish_date'].apply(pd.to_datetime)
```

* Рассчитать прирост студентов на каждом курсе в каждом месяце за всю историю (каждый месяц в диапазоне от марта 2016 до июля 2019 включительно). Считать дату начала прохождения курса студентом по дате начала первой домашней работы.

```python
grouped_months_df = data_df[data_df.is_homework == True]\
                        .groupby(['title', pd.Grouper(key='start_date', freq='M')])\
                        ['student_id'].nunique().reset_index()\
                        .rename(columns={'student_id': 'students_count'})
grouped_months_df
```

* На основании первого пункта построить line-graph с приростом студентов в каждом месяце для каждого курса. 15 графиков. Графики должны иметь заголовки, оси должны быть подписаны. Ticks нужно развернуть так, чтобы они были читаемы.

```python
fig, axes = plt.subplots(5, 3, squeeze=False, figsize=(30, 20))

for ax, t in zip(axes.ravel(), grouped_months_df.title.unique()):
    temp_df = grouped_months_df[grouped_months_df.title.eq(t)]\
                .sort_values(by='start_date')
    ax.set_title(t, fontsize=16)
    
    
    plt.tight_layout()
    
    temp_df.plot.line(x='start_date', y='students_count', ax=ax, legend=None)
    
    x_labels = [pd.to_datetime(str(i)).strftime('%Y-%m') for i in temp_df.start_date]
    
    ax.set_xlabel('Dates', fontsize=16)
    ax.set_ylabel('Students count', fontsize=16)
    
plt.show()
```

```python
module_grouped = data_df.groupby(['title', 'module_number', 'module_title', 'student_id']).agg({'start_date': 'min', 'finish_date': 'max'}).reset_index()
module_grouped.loc[:, 'duration'] = (module_grouped.finish_date - module_grouped.start_date) / pd.Timedelta(days=1)
module_grouped = module_grouped.groupby(['title', 'module_number', 'module_title']).agg({'duration': ['min', 'max', 'mean', 'median']}).reset_index()
```

* На основании первого пункта построить line-graph с медианным временем прохождения каждого модуля для каждого курса. 15 графиков. Графики должны иметь заголовки

```python
fig, axes = plt.subplots(5, 3, figsize=(30, 20))

for ax, t in zip(axes.ravel(), module_grouped.title.unique()):
    temp_df = module_grouped[module_grouped.title.eq(t)]\
                .dropna()\
                .sort_values(by='module_number')
    ax.set_title(t, fontsize=16)
    
    
    plt.tight_layout()
    
    temp_df.plot.line(x='module_number', y=('duration', 'median'), ax=ax, legend=None)
    
    ax.set_xlabel('Module number', fontsize=16)
    ax.set_ylabel('Duration median', fontsize=16)
    
plt.show()
```

* Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе. Формула: отношение количества студентов, приступивших к выполнению домашнего задания в этом модуле (если дз в модуле несколько, то считать по первому дз в модуле), к количеству студентов, сдавших задание в предыдущем модуле (если дз в модуле несколько, то считать по последнему дз в модуле).

```python
conversion = data_df[data_df.is_homework == True].groupby(['title', 'module_number', 'module_title', 'student_id'])\
                    .agg({'start_date': 'min', 'finish_date': 'max'}).reset_index()\
                    .groupby(['title', 'module_number', 'module_title'])[['start_date', 'finish_date']].count().reset_index()\
                    .rename(columns={'start_date': 'started', 'finish_date': 'finished'})
```

```python
conversion['finished_previous'] = conversion.groupby('title')['finished'].shift()
```

```python
conversion['conv_rate'] = (conversion.started / conversion.finished_previous).fillna(1)
```

* Постройте bar-chart, отражающий конверсию перехода студентов из одного модуля в другой на каждом курсе. График должен иметь заголовок. Ticks нужно развернуть так, чтобы они были читаемы

```python
len(conversion.title)
```

```python
for t in conversion.title.unique():
    temp_df = conversion[conversion.title.eq(t)]\
                .sort_values(by='module_number')
    
    fig, ax = plt.subplots()
    ax.set_title(t, fontsize=16)
        
    
    plot = sns.barplot(orient='v', y='conv_rate', x='module_title', data=temp_df)
    plot.set_xticklabels(plot.get_xticklabels(), rotation=30, horizontalalignment='right')

    ax.set_xlabel(None)
    ax.set_ylabel('Conversion rate', fontsize=16)
    
plt.show()
```

Прозрачность должна стоять на отметке 0.1. На графики должна быть нанесена линия медианы конверсии для каждого курса. У медианы должен быть свой цвет. Рамок у графиков быть не должно ⭐

```python
# Создадим несколько вспомогательных объектов для группировки bar-chart по сферам, создадим цветовую палитру и список меток легенды.

fields = {v: k for k, v in enumerate(courses_df['field'].unique())}
```

```python
color_map = conversion.merge(courses_df, how='left', on='title').groupby(['title', 'field']).nunique().drop(columns=['title', 'field']).reset_index().iloc[:, :2]
color_map['color'] = color_map.field.apply(lambda x: colors[fields[x]])
color_map[['title', 'color']].to_dict()
```
