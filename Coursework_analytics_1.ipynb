```python
data_df = progresses_phases_df.merge(progresses_df, how='left', left_on='progress_id', right_on='id').drop('id', axis=1) 
data_df.head()
```

```python
data_df = data_df.merge(courses_df, how='left', left_on='course_id', right_on='id').drop('id', axis=1)
data_df.head()
```

```python
data_df = course_contents_df.merge(data_df, how='outer', on=['course_id', 'module_number', 'lesson_number'])
```

```python
data_df.head(1)
```

---
Выполняем задания раздела 1.

---

```python
# Общее количество курсов в датасете

courses_number = data_df.course_id.nunique()
courses_number
```

```python
# Количество модулей на каждом курсе

modules_count_df = data_df.groupby(['title'])\
                    ['module_number'].nunique()\
                    .reset_index()\
                    .sort_values(by='module_number', ascending=False)\
                    .rename(columns={'module_number': 'modules_count'})\
                    .reset_index(drop=True)
modules_count_df
```

```python
# Количество уроков в каждом модуле на каждом курсе

lessons_count_df = data_df.groupby(['title', 'module_number'])['lesson_number'].nunique().reset_index().rename(columns={'lesson_number': 'lessons_count'})
lessons_count_df
```

```python
# медианное количество уроков в модуле на каждом курсе

lessons_median_df = lessons_count_df.groupby(['title'])['module_number'].agg('median').reset_index().rename(columns={'module_number': 'lessons_median'})
lessons_median_df
```

```python
# количество учеников на каждом курсе

student_count_by_course = data_df.groupby(['title'])['student_id'].nunique().reset_index().rename(columns={'student_id': 'students_count'})
student_count_by_course.sort_values(by='students_count', ascending=False)
```

---
Для работы с датасетом студентов необходимо подготовить данные:

- удалить пустые значения дней рождения;
- отбросить явно неправильные даты (более 100 лет назад и будущие);
- превратить столбец с днями рождений в формат даты;
- добавить столбец возраста путем вычитания даты рождения из текущей даты.

---

```python
students_df = students_df.dropna(subset=['birthday'])
students_df
```

```python
students_df.fillna(value='Unknown', inplace=True)
```

```python
students_df.info()
```

```python
students_real_bithdays = students_df[(students_df.birthday > '1900-01-01') & (students_df.birthday < '2020-01-01')]
```

```python
students_real_bithdays.loc[:, 'birthday'] = students_real_bithdays.loc[:, 'birthday'].apply(pd.to_datetime)
```

```python
students_real_bithdays.loc[:, 'age'] = pd.to_datetime('2020-08-24') - students_real_bithdays['birthday']

students_real_bithdays.loc[:, 'age'] = students_real_bithdays.loc[:, 'age'] / timedelta(days=365)
```

```python
students_real_bithdays
```

```python
# минимальный, максимальный, средний, медианный возраст студентов

age_statistics = students_real_bithdays.groupby('id')['age'].agg([min, max, 'mean', 'median'])
age_statistics
```

```python
# минимальный, максимальный, средний, медианный возраст студентов на каждом курсе

student_courses_df = students_real_bithdays.merge(progresses_df, how='left', left_on='id', right_on='student_id', suffixes=('_stud', '_prog')) 
student_courses_df = student_courses_df.merge(courses_df, how='left', left_on='course_id', right_on='id')
student_courses_df.drop(columns=['id_prog', 'student_id', 'course_id', 'id', 'field'], inplace=True)
student_courses_df.dropna(inplace=True)
student_courses_df
```

```python
ages_by_courses = student_courses_df.groupby('title')['age'].agg([min, max, 'mean', 'median'])
ages_by_courses
```
